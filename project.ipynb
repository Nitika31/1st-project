```python
from ultralytics import YOLO

# Load classification model
model = YOLO('yolov8n-cls.pt')

# Train - NO YAML FILE NEEDED
results = model.train(
    data='dataset',  # <- Point to ROOT folder (dataset/)
    epochs=25,
    imgsz=224,
    batch=16,
    project='cattle_buffalo_cls',
    name='v1'
)
```
